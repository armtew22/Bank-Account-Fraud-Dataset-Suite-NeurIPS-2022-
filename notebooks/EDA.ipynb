```python
# ordinal encoding for Dtype object values:

from sklearn.preprocessing import OrdinalEncoder

# Display value counts before encoding
for column in ['payment_type', 'employment_status', 'housing_status']:
    print(f"Value counts for {column} before encoding:")
    print(base_df[column].value_counts())
    print()

# Apply ordinal encoding
encoder = OrdinalEncoder()
base_df[['payment_type', 'employment_status', 'housing_status']] = encoder.fit_transform(
    base_df[['payment_type', 'employment_status', 'housing_status']]
)

```

```python
# Display value counts after encoding
for column in ['payment_type', 'employment_status', 'housing_status']:
    print(f"Value counts for {column} after encoding:")
    print(base_df[column].value_counts())
    print()
```

```python
# dropping insignificant data points
base_df = base_df.drop(columns=['device_os', 'source'])
```

```python
base_df.describe()
# vals with -1 for missing: 'prev_address_months_count', 'current_address_months_count', 'bank_months_count', 
```

```python
columns = ['prev_address_months_count', 'current_address_months_count', 'bank_months_count']

for column in columns:
    count = (base_df[column] == -1.0).sum()
    print(f"Column '{column}': -1.0 appears {count} times")
```

```python
# Create 'has_prev_address' column before replacing values
base_df['has_prev_address'] = (base_df['prev_address_months_count'] != -1.0).astype(int)

# Create 'has_prev_bank_acc' column before replacing values
base_df['has_prev_bank_acc'] = (base_df['bank_months_count'] != -1.0).astype(int)

# Replace -1.0 with the mean value for the specified columns
base_df['prev_address_months_count'] = base_df['prev_address_months_count'].replace(-1.0, base_df['prev_address_months_count'].mean())
base_df['bank_months_count'] = base_df['bank_months_count'].replace(-1.0, base_df['bank_months_count'].mean())

# Print value counts for the created columns
print("Value counts for 'has_prev_address':")
print(base_df['has_prev_address'].value_counts())
print()

print("Value counts for 'has_prev_bank_acc':")
print(base_df['has_prev_bank_acc'].value_counts())

```
